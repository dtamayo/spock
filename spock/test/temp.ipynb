# Creates a small testing dataset of systems for quick sanity checks

```python
import pandas as pd
```

```python
datapath = '../../../cleanData/csvs/resonant/'
```

We can load the data based on datapath

```python
initial = pd.read_csv(datapath+'clean_initial_conditions.csv')
label = pd.read_csv(datapath+'clean_labels.csv')
```

```python
new = pd.DataFrame.join(initial, label['Stable'])
```

We can pick a set of 100 simulations that would be in the testing set, and try and make sure there is a decent mix of stable/unstable

```python
new['Stable'].iloc[-2000:-1900].value_counts()
```

```python
new = new.iloc[-2000:-1900]
```

```python
#clean up the datasheet
new = new.drop(columns = ['Unnamed: 0'])
```
